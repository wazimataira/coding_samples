Stable DiffusionXL model

```python
from PIL import Image
import torch
from torch import autocast
from diffusers import StableDiffusionXLPipeline,EulerAncestralDiscreteScheduler

model_id="Linaqruf/animagine-xl"#@param{type:"string"}

pipe=StableDiffusionXLPipeline.from_pretrained(model_id,torch_dtype=torch.float16, variant="fp16", use_safetensors=True
)


pipe.scheduler = EulerAncestralDiscreteScheduler.from_config(pipe.scheduler.config)
pipe.enable_model_cpu_offload()

default_prompt = " masterpiece, best quality"
negative_prompt = "lowres, bad anatomy, bad hands, text, error, missing fingers, extra digit, fewer digits, cropped, worst quality, low quality, normal quality, jpeg artifacts, signature, watermark, username, blurry"
```

```python
prompt="silver hair, short pants, beach, smile, coll girl,bust shot,face center,clearly,midium bust,short hair,bikini,looking at viewer, upper body,normal body,sexy"#@param{type:"string"}
prompt=default_prompt+prompt
image = pipe(
    prompt,
    negative_prompt=negative_prompt,
    width=1024,
    height=1024,
    guidance_scale=12,
    target_size=(1024,1024),
    original_size=(4096,4096),
    num_inference_steps=100
    ).images[0]

name="pics4"#@param{type:"string"}
name="./pics/"+name+".png"
image.save(name)
```

```python
import gc

if "pipe" in globals():
    del pipe
if "image" in globals():
    del image
torch.cuda.empty_cache()
gc.collect()
```
